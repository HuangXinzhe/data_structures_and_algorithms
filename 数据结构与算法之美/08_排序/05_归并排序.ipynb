## 归并排序
把数组从中间分成前后两部分，然后对前后两部分分别排序，再将排好序的两部分合并在一起


```python
def merge(left_arr, right_arr):
    result = []
    while left_arr and right_arr:
        if left_arr[0] <= right_arr[0]:
            result.append(left_arr.pop(0))
        else:
            result.append(right_arr.pop(0))
    
    while left_arr:
        result.append(left_arr.pop(0))

    while right_arr:
        result.append(right_arr.pop(0))
    
    return result
    

def merge_sort(arr):
    if len(arr) < 2:
        return arr
    mid = len(arr) // 2
    left_arr = merge_sort(arr[:mid])
    right_arr = merge_sort(arr[mid:])
    return merge(left_arr, right_arr)
```

```python
result = merge_sort([10, 1, 9, 2, 5, 3, 8, 4, 7, 6])
print(result)
```

1. 归并排序是稳定的排序算法
2. 归并排序最好情况、最坏情况，还是平均情况，时间复杂度都是 O(nlogn)。
3. 归并排序的空间复杂度是O(n)
