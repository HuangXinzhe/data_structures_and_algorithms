## Selection Sort


 O(1) O(n2)
### 
- 

```python
def selection(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i+1,len(arr)):
            if arr[min_index] > arr[j]:
                min_index = j
        arr[i],arr[min_index] = arr[min_index],arr[i]
    return arr
```

```python
result = selection([5,4,3,2,1])
print(result)
```
