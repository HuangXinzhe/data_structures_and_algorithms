## 插入排序（Insertion Sort）
遍历数组，找到数据应该插入的位置将其插入即可。

将数组中的数据分为两个区间，已排序区间和未排序区间。初始已排序区间只有一个元素，就是数组的第一个元素。插入算法的核心思想是取未排序区间中的元素，在已排序区间中找到合适的插入位置将其插入，并保证已排序区间数据一直有序。重复这个过程，直到未排序区间中元素为空，算法结束。

插入排序也包含两种操作，一种是元素的比较，一种是元素的移动

```python
def insertion(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j-=1
        arr[j+1] = key
    return arr
```

```python
result = insertion([12, 11, 13, 5, 6])
print(result)
```

### 要点
1. 插入排序是原地排序算法
2. 插入排序是稳定的排序算法
3. 插入排序的最好时间复杂度是O(n),最坏时间复杂度是O(n^2),平均时间复杂度是O(n^2)
