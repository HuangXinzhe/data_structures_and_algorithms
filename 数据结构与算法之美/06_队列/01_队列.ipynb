## 队列
- 先进先出
- 队尾进，队头出


- 用数组实现的队列叫作顺序队列
- 用链表实现的队列叫作链式队列

队列需要两个指针：一个是 head 指针，指向队头；一个是 tail 指针，指向队尾。  

基于数组的方法实现队列
- 入队操作，将数据插入到 tail 指针指向的位置
- 出队操作，将 head 指针指向的数据删除
- 当入队操作发现队列已满时，进行数据搬移



基于链表的队列实现方法  
入队时，tail->next= new_node, tail = tail->next；出队时，head = head->next


### 循环队列
- 队列的出队和入队操作时间复杂度都是 O(1)
- 队列的出队操作平均时间复杂度也是 O(1)
- 队列的出队操作平均时间复杂度是 O(1) 的前提是，没有数据搬移的情况。而在入队操作中，数据搬移是必须的，所以，入队操作的时间复杂度是有可能均摊到 O(n) 的。
- 循环队列的队满条件是 (tail+1)%n=head，队空条件是 head=tail
