  
f(n)=f(n-1)+1 f(1)=1

```python
# 
def recursion(input):
    if input == 1:
        return 1
    return recursion(input - 1) + input
```

```python
result = recursion(5)
print(result)
```

```python
# 
def walk(input):
    if input == 1:
        return 1
    if input == 2:
        return 2
    return walk(input - 1) + walk(input - 2)
```

```python
result = walk(5)
print(result)
```

```python
def walk2(input):
    if input == 1:
        return 1
    if input == 2:
        return 2
    first = 1
    second = 2
    third = 0
    for _ in range(3, input+1):
        third = first + second
        first = second
        second = third
    return second
```

```python
result = walk2(5)
print(result)
```
